```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
```

```python
loss = pd.read_csv("metrics/loss.csv")
roc = pd.read_csv("metrics/roc.csv")
pr = pd.read_csv("metrics/pr.csv")
```

```python
plt.plot(loss["Iteration"], loss["Loss"])
plt.title("Loss across iterations")
plt.xlabel("Iteration")
plt.ylabel("Loss")
```

```python
plt.plot(pr["FPR"], pr["TPR"])
plt.title("ROC Learning Curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
```

```python
plt.plot(roc["recall"], roc["precision"])
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
```
